# Programa de calificaciones

Este programa calcula el promedio de tes calificaciones y evalua si el alumno **Aprobo** o **No aprobo**

## Definicion de variables

```python
calif1 = 0.0
calif2 = 0.0
calif3 = 0.0
promedio = 0.0
```

# Entrada de datos

```python
calif1 = float(input("Introduce la calificacion 1"))
calif2 = float(input("Introduce la calificacion 2"))
calif3 = float(input("Introduce la calificacion 3"))
```

# Proceso: Calculo promedio

```python
promedio = (calif1 + calif2 + calif3) / 3
```

# Comparacion con una Estructura selectiva doble

```python
if promedio >= 7.0:
    print("Aprobado")
else:
    print("No aprobado")
```
